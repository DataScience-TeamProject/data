# **Ⅱ  비자 유형에 따른 외국인 체류 요인**

# **1. 라이브러리 및 데이터 불러오기**

**사용한 데이터**

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

#범례를 나타내기 위한 라이브러리
import matplotlib.patches as mpatches 
import matplotlib.lines as mlines 
```

```python
#csv 파일 불러오기
for_stay = pd.read_csv('법무부_외국인체류데이터_20191231.csv', encoding='cp949')
for_stay.head()
```

# **2. 데이터 전처리**

**데이터 전처리 과정**

```python
#NaN값 0으로 채우기
for_stay = for_stay.fillna(0)

#대륙, 국적, 성별 칼럼 제거
for_stay = for_stay.drop(['대륙', '국적', '성별'], axis=1)

#값을 정수형(int)으로 변환
for_stay = for_stay.astype(int)

#컬럼명 정제
for_stay.columns = ['문화예술','유학','기술연수','일반연수','취재', '종교', '주재', '기업투자', '무역경영',
              '구직', '교수', '회화강사', '연구', '기술지도', '전문직업', '예술흥행', '특정활동', 
              '비전문취업', '선원취업', '방문동거', '거주', '동반', '영주', '결혼이민', '기타_1', '관광취업', '방문취업', '기타_2']

#[기타_1]과 [기타_2]를 [기타]로 병합
for_stay['기타'] = for_stay['기타_1']+for_stay['기타_2']
for_stay = for_stay.drop(['기타_1', '기타_2'], axis=1)

for_stay.head()
```

```python
#컬럼명 확인
for_stay.columns
```

```python
#리스트에 컬럼 저장
columns = ['문화예술', '유학', '기술연수', '일반연수', '취재', '종교', '주재', '기업투자', '무역경영', '구직',
       '교수', '회화강사', '연구', '기술지도', '전문직업', '예술흥행', '특정활동', '비전문취업', '선원취업',
       '방문동거', '거주', '동반', '영주', '결혼이민', '관광취업', '방문취업', '기타']

#각 컬럼의 합계 도출
sum = for_stay[columns].sum()
sum
```

```python
#새 데이터프레임 생성
for_stay_new = pd.DataFrame()

#컬럼명 지정, 값은 sum으로 지정
for_stay_new['응답수'] = sum

#인덱스를 '체류요인' 컬럼으로 변경
for_stay_new = for_stay_new.rename_axis('체류요인').reset_index()
for_stay_new.head(10)
```

```python
#비자 유형별로 그룹화해 컬럼 생성
for_stay_new = for_stay_new.assign(비자유형 = ['취업불가비자']*9 + ['전문인력 취업비자']*8 + ['단순노무 취업비자']*2
                                   + ['취업불가비자'] + ['업무무관 취업가능비자'] + ['취업불가비자']
                                   + ['업무무관 취업가능비자']*2 + ['단순노무 취업비자']*3)

#'비자유형' 컬럼을 기준으로 오름차순 정렬
for_stay_new = for_stay_new.sort_values(by = ['비자유형'], ascending = [True])

for_stay_new
```

# **3. 시각화**

**시각화 과정**


1) 비자 유형에 따른 외국인 체류 요인
- 27개의 체류 요인을 네 가지의 비자 유형으로 그룹화하였다.
- 취업이 가능한 세 유형(단수노무 취업비자, 업무무관 취업가능비자, 전문인력 취업비자)과 취업이 불가한 비자(취업불가비자) 유형으로 구분하였다.

```python
#그래프 크기 설정
plt.rcParams['figure.figsize'] = (12, 10)

#폰트 설정
plt.rc('font', family='NanumBarunGothic')
```

```python
#색상 지정
cmap = plt.get_cmap('Set3')
colors = cmap([7,7,7,7,7,10,10,10,4,4,4,4,4,4,4,4,2,2,2,2,2,2,2,2,2,2,2])

#전체 비율 중 1 이상인 비율만 표기
def customPct(pct):
  return('%1.1f%%'%pct) if pct >=1 else ''

#원그래프 설정
plt.title('비자 유형에 따른 외국인 체류 요인', fontsize = 20)
plt.pie(for_stay_new['응답수'], autopct = customPct, colors = colors,
        startangle=90, counterclock=False)

#범례 지정
legend_a = mpatches.Patch(color = 'pink', label = '단순노무 취업비자')
legend_b = mpatches.Patch(color = 'palegreen', label = '업무무관 취업가능비자')
legend_c = mpatches.Patch(color = 'steelblue', label = '전문인력 취업비자')
legend_d = mpatches.Patch(color = 'plum', label = '취업불가비자')

#범례 출력
plt.legend(handles = [legend_a, legend_b, legend_c, legend_d], loc = (1, 0.5))

plt.show()
```

2) 비자 유형이 '단순노무 취업비자'인 항목 강조

```python
#전체 비율 중 1 이상인 비율만 표기
def customPct(pct):
  return('%1.1f%%'%pct) if pct >=1 else ''

#비자 유형이 '단순노무 취업비자'인 항목 강조
explode = ([0.05]*5 + [0]*22)

#원그래프 그리기
plt.title('비자 유형에 따른 외국인 체류 요인', fontsize = 20)
plt.pie(for_stay_new['응답수'], autopct = customPct, colors = colors,
        startangle=90, counterclock=False, explode = explode)

#범례 지정하기
legend_a = mpatches.Patch(color = 'pink', label = '단순노무 취업비자')
legend_b = mpatches.Patch(color = 'palegreen', label = '업무무관 취업가능비자')
legend_c = mpatches.Patch(color = 'steelblue', label = '전문인력 취업비자')
legend_d = mpatches.Patch(color = 'plum', label = '취업불가비자')

#범례 출력
plt.legend(handles = [legend_a, legend_b, legend_c, legend_d], loc = (1, 0.5))

plt.show()
```
