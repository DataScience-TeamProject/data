```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
with open('example_txt_data.txt', 'r', encoding='utf-8') as f:
    text = f.read()

okt = Okt()
nouns = okt.nouns(text) # 명사만 추출

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외

c = Counter(words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
c
```

```python
#한글 폰트 설정
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
#한글로 워드클라우드 시각화
wc = WordCloud(font_path='NanumBarunGothic', background_color='white', width=400, height=400, scale=2.0, max_font_size=250)
cloud = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(cloud)
```

```python
#파일로 저장하기
wc.to_file('예시그림.png')
```

```python
#사진 모양대로 워드클라우드 만들기
img = Image.open('blackheart1.png')
img_array = np.array(img)

wc = WordCloud(font_path='NanumBarunGothic', background_color='white', width=400, height=400, scale=2.0, max_font_size=250, mask=img_array)
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
```

```python
#파일로 저장하기
wc.to_file('예시그림하트.png')
```
